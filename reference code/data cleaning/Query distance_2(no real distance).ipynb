```python
import pandas as pd
import numpy as np
import time

df = pd.read_csv('Merged_data_v2.csv')
df
```

```python
#haversine_distance（直線大圓距離）（km)
R = 6371  #地球半徑

#轉換為弧度
lat1 = np.radians(df['from_latitude'])
lon1 = np.radians(df['from_longitude'])
lat2 = np.radians(df['to_latitude'])
lon2 = np.radians(df['to_longitude'])

dlat = lat2 - lat1
dlon = lon2 - lon1

a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
c = 2 * np.arcsin(np.sqrt(a))

df['haversine_distance'] = R * c
```

```python
#manhattan_distance（l1距離）

# 緯度差（南北向）
dlat = np.abs(lat2 - lat1)
lat_dist = R * dlat

# 經度差（東西向），取兩地平均緯度作為參考
avg_lat = (lat1 + lat2) / 2
dlon = np.abs(lon2 - lon1)
lon_dist = R * dlon * np.cos(avg_lat)

df['manhattan_distance'] = lat_dist + lon_dist
```

```python
df.info()
```

```python
df = df[df['volume']!='#NUM!']
```

```python
df['log_cost'] = np.log1p(df['cost'])
df['log_weight'] = np.log1p(df['weight'])
df['log_Hdis'] = np.log1p(df['haversine_distance'])
df['log_Mdis'] = np.log1p(df['manhattan_distance'])
df['volume'] = df['volume'].astype(float)
df['log_volume'] = np.log1p(df['volume'])
df['log_TVP'] = np.log1p(df['total_vendor_price'])
df['log_TVP/log_weight'] = df['log_TVP'] / df['log_weight']
df['across_state'] = (df['from_state'] != df['to_state'])
del df['Days to Ship']
df
```
