## 建立函式把所有檔案搞完

```python
def dealing_tvp_and_volume(df: pd.DataFrame, tvp):
    duplicated_ids = df[df.duplicated('Shipment ID', keep = False)].sort_values('Shipment ID')
    group_cost_sum = duplicated_ids.groupby('Shipment ID')['weight'].transform('sum')
    duplicated_ids['volume'] = (duplicated_ids['volume'] * duplicated_ids['weight'] / group_cost_sum).round(2)
    if tvp:
        duplicated_ids['total_vendor_price'] = (duplicated_ids['total_vendor_price'] * duplicated_ids['weight'] / group_cost_sum).round(2)
    df.update(duplicated_ids)
    if tvp:
        df['log_TVP'] = np.log1p(df['total_vendor_price'])
        df['log_TVP/log_weight'] = df['log_TVP'] / df['log_weight']
    df['log_volume'] = np.log1p(df['volume'])
    return df
```

```python
from pathlib import Path
dflist = [f.name for f in Path('.').glob('*.csv')]
dflist
```

```python
dflist = [f for f in dflist if f[:4]=='Data']
dflist
```

```python
for name in dflist:
    df = pd.read_csv(name)
    if name[6] == 3:
        pass
    elif name[6] == 6 or name[6] == 8:
        df = dealing_tvp_and_volume(df, 1)
    else:
        df = dealing_tvp_and_volume(df, 0)
    df.to_csv(name, index=False)
    print(f'{name} completed.')
```
